```python
from konlpy.corpus import kolaw
```

```python
text = kolaw.open('constitution.txt').read()
```

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
noun_text = okt.nouns(text)
```

```python
from collections import Counter
counts = Counter(noun_text)
```

```python
data = counts.most_common(500)
```

```python
data = dict(data)
```

```python
wc = WordCloud(font_path = '/usr/share/fonts/truetype/nanum/NanumMyeongjoBold.ttf', background_color= 'pink').generate_from_frequencies(data)


mpl.figure(figsize=(8,6))
mpl.imshow(wc)
mpl.axis('off')
mpl.show()
```

```python
from konlpy.tag import Hannanum
```

```python
text = kolaw.open('constitution.txt').read()
```

```python
han = Hannanum()
text_noun = han.nouns(text)
```

```python
stopwords = set(['제','월','일','조','수','때','그','이','바','및','안','정','위','장'])
```

```python
noun_law = [each_word for each_word in text_noun if each_word not in stopwords]
```

```python
noun_law
```

```python
count = Counter(noun_law)
data = count.most_common(500)
```

```python
data = dict(data)
```

```python
korea = np.array(Image.open('/content/drive/MyDrive/파이썬AI프로젝트 202307/3.데이터 분석/mask_korea.jpg'))
```

```python
wc = WordCloud(background_color= 'black',mask= korea, font_path ='/usr/share/fonts/truetype/nanum/NanumGothic.ttf').generate_from_frequencies(data)

mpl.figure(figsize=(10,13))
mpl.imshow(wc)
mpl.axis('off')
mpl.show()
```
